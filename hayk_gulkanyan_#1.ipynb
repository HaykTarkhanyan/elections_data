The average number of people registered at the same address #1


```python
import pandas as pd

df = pd.read_parquet("elections.parquet")
df
```

```python
count_of_members = df.groupby('hasce').size().reset_index(name='count')

count_of_members
```

```python
df['haeranun'] = df['haeranun'].astype(str).str[:-1]
df['haeranun'] = df['haeranun'].astype(str).str.replace(r'յ$', '', regex=True)
```

```python
boys = df['haeranun']

girls = df[~df['anun'].isin(boys)]['anun'].reset_index(drop=True)
girls
```

```python

girls_df = df[~df['anun'].isin(boys)].reset_index(drop=True)
boys_df = df[df['anun'].isin(boys)].reset_index(drop=True)


girls_count_by_address = girls_df.groupby('hasce').size().reset_index(name='count')
boys_count_by_address = boys_df.groupby('hasce').size().reset_index(name = 'count')
boys_count_by_address
```
